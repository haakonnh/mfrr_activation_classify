```python


import pandas as pd

df = pd.read_csv('../data/raw/BalanceMarket_2025_NO1_EUR_None_MW.csv', delimiter=';')
df2 = pd.read_csv('../data/raw/BalanceMarket_2024_NO1_EUR_None_MW.csv', delimiter=';')
```

```python
df = pd.concat([df2, df], ignore_index=True)
total_entries = len(df)
total_entries
```

```python
nonzero_down = (df['NO1 Activated Down Volume (MW)'] != 0).sum()
nonzero_up = (df['NO1 Activated Up Volume (MW)'] != 0).sum()
both_zero = ((df['NO1 Activated Down Volume (MW)'] == 0) & (df['NO1 Activated Up Volume (MW)'] == 0)).sum()
both_nonzero = ((df['NO1 Activated Down Volume (MW)'] != 0) & (df['NO1 Activated Up Volume (MW)'] != 0)).sum()
nonzero_down, nonzero_up, both_zero, both_nonzero
```

```python
sum([nonzero_down, nonzero_up, both_zero, both_nonzero]) - 2 * both_nonzero, total_entries
```

```python
test_df_last20percent = df.tail(int(0.2 * total_entries))
non_zero_down_test = (test_df_last20percent['NO1 Activated Down Volume (MW)'] != 0).sum()
non_zero_up_test = (test_df_last20percent['NO1 Activated Up Volume (MW)'] != 0).sum()
both_zero_test = ((test_df_last20percent['NO1 Activated Down Volume (MW)'] == 0) & (test_df_last20percent['NO1 Activated Up Volume (MW)'] == 0)).sum()
both_nonzero_test = ((test_df_last20percent['NO1 Activated Down Volume (MW)'] != 0) & (test_df_last20percent['NO1 Activated Up Volume (MW)'] != 0)).sum()
non_zero_down_test, non_zero_up_test, both_zero_test, both_nonzero_test, sum([non_zero_down_test, non_zero_up_test, both_zero_test, both_nonzero_test]) - 2 * both_nonzero_test
```
